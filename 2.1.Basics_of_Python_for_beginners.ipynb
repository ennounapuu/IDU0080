###Python Data Structures:
Python provides different types of data structures, and referred to as sequence types.

**List**

A list is an ordered sequence which contains a number of items enclosed by a square bracket separated by commas. Python allows list of single type as well as mixed types. Lists are mutable, meaning the items in the list can be modified.

````
dataList = []                         # Empty list
dataList = [1, 2, 3]                  # List of integers
dataList = [1, "Hello", 2, "World"]   # List of mixed data types
````

To access the elements of the list, we can use index operator [ ]. Index, in Python, starts from 0. We can use negative indexs as well in python. The index -1 refers to the last item, -2 to the second last and so on.

````
dataList = [1, 2, 3, 4]
print("Second element is: ", dataList[1])
print("The last element is: ", dataList[-1])
````

Python offers multiple list methods that can be used to work with lists with ease. Some of the commonly used methods are:
- append(): add an element to the list
- insert(): insert an element to the list
- remove(): remove element from the list
- reverse(): reverse a list
- sort(): sort the elements of the list
- clear(): remove all elements of a list
- len(): returns size of list

Lets look at some examples.

````
dataList = [1, 2, 3, 4]

# change item value
print("change item value")
dataList[1] = 9
print(dataList)

# check if item exists in list
print("\ncheck if item exists")
if 3 in dataList:
  print("3 exists")
  
# length of list
print("\nlength of list")
print(len(dataList))

#looping through list
print("\nlooping through list")
for x in dataList:
  print(x)

# add items
print("\nadd item to list")
dataList.append(5)
print(dataList)

# remove item
print("\nremove item from list")
dataList.remove(5)
print(dataList)

# sorting list
print("\nsorting")
print("Before sort:", dataList)
dataList.sort()
print("After sort:", dataList)

# clear list
print("\nclear list")
dataList.clear()
print(dataList)
````

**Dictionary**

A dictionary is an unordered collection of items, with element as key:value pair. Other compound data stuctures have only value as element but dictionary has key:value pair. It can have mixed type elements, including collection types.

````
dataDict = {}                            # Empty dictionary
dataDict = {1: "apple", 2: "ball"}       # Dictionary with int keys
dataDict = {"a": "apple", 1: [2, 3, 5]}  # Dictionary with mixed keys
````

Accessing the value from a dictionary is done using keys.

````
dataDict = {"name": "apple", "category": "fruit", "price": 105}
print(dataDict["name"])
print(dataDict["price"])
````

Like other data structures, Python also provides some methods for working with dictionaries. Some of these are:

- get(): returns value of key
- keys(): returns all keys
- values(): returns all values
- update(): update dictionary
- clear(): remove all elements
- len(): returns size of dictionary

Lets look at some of these methods.

````
dataDict = {"name": "apple", "category": "fruit", "price": 105}

# accessing data
print("accessing data")
print(dataDict["name"])
print(dataDict.get("name"))


# change value
print("\nchange value")
dataDict["name"] = "banana"
print(dataDict)

# looping through dictionary
print("\nlooping through dictionary")
print("- printing keys")
for x in dataDict:
  print(x)
print("- printing values")
for x in dataDict:
  print(dataDict[x])

# get all values
print("\nget all values")
print(dataDict.values())

# check if key exists
print("\ncheck if key exists")
if "name" in dataDict:
  print("name key exists")
  
# dictionary length
print("\nlength of dictionary")
print(len(dataDict))

# adding items
print("\nadding item")
dataDict["color"] = "yellow"
print(dataDict)

# clear dictionary
print("\nclear dictionary")
dataDict.clear()
print(dataDict)
````

###Flow Control

**if...else**

We use if...else statement when we want to run different actions on different conditions. 

````
n = 5

if n < 10:
  print("Less than 10")
elif n > 10:
  print("Greter than 10")
else:
  print(n)
````

There can be no elif parts and the else part can be omitted too. 

Unlike most other programming languages where they use { } to denote block of code, Python uses indentation.

````
if False:
  print("First inside if")
  print("Second inside if")

print("Outside if")
````

**while loop**

Like every other programming languages, while loop is used to loop over a block of statements as long as the test condition is true.

````
n = 1

while n < 10:
  print(n)
  n = n + 1
````

**for loop**

In Python, for loop can be used to iterate over list, tuple, string and such other sequence of iterable objects.

````
numbers = [1, 2, 3, 4, 5]

for n in numbers:
  print(n)
````

The "in" operator returns a True if value is found in the sequence otherwise returns a False.

**break statement**

A break statement terminates the loop which contains the break statement. No code below the break statement is executed and the flow is immediately thrown outside the loop.

````
numbers = [1, 2, 3, 4, 5]

for x in numbers:
  if x == 3:
    break
  print(x)
  
print("Outside")
````

**continue statement**

The continue statement skips the rest of the statments below the continue statement, and continues to the next loop. It does not terminate the loop but continues with next iteration.

````
numbers = [1, 2, 3, 4, 5]

for x in numbers:
  if x == 3:
    continue
  print(x)
  
print("Outside")
````

###Object Oriented Programming in Python
In Python, everything is an object. This includes integers, floats, functions and classes. Lets look at how we can create classes and objects.

###Class and Objects
A class is a user defined data type, which consists of its own data and methods. A class is regarded as a blueprint of object.

**Defining a Class**



````
class FruitClass:
  name = "apple"
  
  def display(self):
    print("This is Fruit Class")
    

print(FruitClass.name)
````

In the above example, we have defined a class with a "name" variable and display() function as its components.

**Creating Objects**



````
class FruitClass:
  name = "apple"
  
  def display(self):
    print("This is Fruit Class")
    
fruitObject = FruitClass()
print(fruitObject.name)
fruitObject.display()
````

We can create objects of the class, and use the object to access the members (variables and functions) of the class.

**Constructors**

We use \__init\__ as constructor in Python classes. This method is called automatically when an object of the class is instantiated.

````
class FruitClass:
  name = "apple"
  
  def __init__(self):
    self.name = "banana"
    
fruitObject = FruitClass()
print(fruitObject.name)
````

In the above example, the default value for "name" is set to "apple", but the constructor sets the "name" to "banana".

Notice the use of "self" parameter in functions and constructors inside the class. But we simply used display() to call the function without the parameter, and it worked. The "self" paramater represents the object itself and whenever an object calls the method, the object itself is passed as the first argument implicitly. In other words, the method is translated from fruitObject.display() to FruitClass.display(fruitObject).

````
import numpy as np

list1 = [1, 2, 3]
print("List: ", list1)

# add array
array2 = array1 + array1
print("Add: ", array2)

# multiply array
array2 = array1 * array1
print("Multiply: ", array2)

# divide array
array2 = array1 / array1
print("Divide: ", array2)

list1 = [1, 2, 3, 4, 5]
print("\nList: ", list1)
# sine function
print("Sine: ", np.sin(list1))

# log
print("Natural Log: ", np.log(list1))
print("Base-2 Log: ", np.log2(list1))
print("Base-10 Log: ", np.log10(list1))

# exp
print("Exponential: ", np.exp(list1))
````

###Pandas
Pandas is a library for Python which provides high-performance manipulation and analysis of data structures.

**Creating Pandas series**

Pandas series are 1-D labeled array which can hold any type of data. Pandas series can be created using lists and numpy arrays.

````
import numpy as np
import pandas as pd

keys = ['a', 'b', 'c']
values = [1, 2, 3]
npArr = np.array(values)
dataDict = {'a': 1, 'b': 2, 'c':3}

ps1 = pd.Series(data=values)
print(ps1)

ps2 = pd.Series(data=values, index=keys)
print(ps2)

ps3 = pd.Series(npArr, keys)
print(ps3)

ps4 = pd.Series(dataDict)
print(ps4)
````

**Creating Pandas DataFrame**

Pandas DataFrame is 2-D data structure. It is similar to Pandas series but in 2-D. Pandas also offers some methods to handle these DataFrames.

````
import numpy as np
import pandas as pd

mat = np.random.randint(1, 10, 12).reshape(3, 4)
rowLabels = ['A', 'B', 'C']
colLabels = ['W', 'X', 'Y', 'Z']

df = pd.DataFrame(data=mat, index=rowLabels, columns=colLabels)
print("Data Frame")
print(df)
````

````
import numpy as np
import pandas as pd

mat = np.random.randint(1, 10, 100).reshape(25, 4)
partialColLabels = ['W', 'X', 'Y', 'Z']
partialDf = pd.DataFrame(data=mat, columns=partialColLabels)

print("Data Frame")
print(partialDf)

print("\nPartial Data Frame")
print(partialDf.head())

print("\nPartial Data Frame Head(10)")
print(partialDf.head(10))

print("\nPartial Data Frame Tail(5)")
print(partialDf.tail(5))
````

**Indexing and Slicing**

Pandas DataFrame can be split as required from the original and sliced into pieces based on the row or column labels or indices.

````
import numpy as np
import pandas as pd

mat = np.random.randint(1, 10, 20).reshape(5, 4)
rowLabels = ['A','B','C','D','E']
colLabels = ['W', 'X', 'Y', 'Z']

df = pd.DataFrame(data=mat, index=rowLabels, columns=colLabels)
print(df)

print("\nSingle  Row - label B")
print(df.loc['B'])

print("\nMulti Rows - labels B and D")
print(df.loc[['B', 'D']])

print("\nSingle  Row - index 2")
print(df.iloc[2])

print("\nMulti Rows - indixes 2 and 3")
print(df.iloc[[2, 3]])
````

**Creating and Deleting rows and columns**

We can also add new rows or columns to the DataFrame. The new row or column can also be obtained from operating on the other rows or columns. Similary we can also delete them as required. 

````
import numpy as np
import pandas as pd

mat = np.random.randint(1, 10, 20).reshape(5, 4)
rowLabels = ['A','B','C','D','E']
colLabels = ['W', 'X', 'Y', 'Z']

df = pd.DataFrame(data=mat, index=rowLabels, columns=colLabels)
print(df)

print("\nAdding new column")
df['X+Y'] = df['X'] + df['Y']
print(df)

print("\nDelete column X+Y")
df = df.drop('X+Y', axis=1)
print(df)

print("\nDelete row A")
df = df.drop('A')
print(df)
````

###Matplotlib
Matplotlib is a 2D plotting library for Python.  It produces plots and makes things easy by providing controls for the plots. It can be used to create scatter plots, histogram, bar charts, spectra, etc.

**Scatter Plot**

We can create scatter plot as follows:

````
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

people = ['Jack', 'John', 'Amy', 'Shun', 'June',
         'May', 'Janet', 'Cherry', 'Ross', 'Chand']
age = [14, 15, 19, 23, 22, 17, 18, 24, 21, 25]
marks = [77, 45, 65, 80, 65, 60, 85, 88, 90, 93]

plt.scatter(age, marks)
plt.show()
````

**Formatting plots**

The plots can be formatted to look as our needs using the controls matplotlib provides us. We can add xlabel, ylabel, grid lines, title, legends, etc. We can also customize the colors of the plots.

````
plt.figure(figsize=(8, 6))
plt.title("Plot of Age vs Marks", fontsize=20)
plt.xlabel("Age (years)", fontsize=16)
plt.ylabel("Marks", fontsize=16)
plt.grid (True)
plt.ylim(0,100)
plt.xticks([i*5 for i in range(12)],fontsize=15)
plt.yticks(fontsize=15)
plt.scatter(x=age, y=marks, c='blue', s=150, edgecolors='k')
plt.legend(['Marks'], loc=2, fontsize=12)
plt.show()
````

**Bar graphs**

Matplotlib can be used to create bar graphs.

````
plt.figure(figsize=(7, 4))
plt.bar(x=np.arange(0, 10), height=age)
plt.xlabel("serial", fontsize=15)
plt.ylabel("Ages", fontsize=15)
plt.show()
````

**Histogram**

Similarly, we can also create histograms.

````
plt.figure(figsize=(7, 5))
plt.hist(marks, color='blue', edgecolor='k')
plt.xlabel("Marks", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
````

**Pandas Dataframe plot**
